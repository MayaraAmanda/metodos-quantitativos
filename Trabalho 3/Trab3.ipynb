### 1. Distribuição Bernoulli

A distribuição de Bernoulli é um caso especial da distribuição binomial, em que um único experimento é conduzido para que o número de observações seja 1. Consideremos um experimento que consiste em uma sequência de ensaios ou tentativas independentes, isto é, ensaios nos quais a probabilidade de um resultado em cada ensaio não depende dos resultados ocorridos nos ensaios anteriores, nem dos resultados nos ensaios posteriores. Em cada ensaio, podem ocorrer apenas dois resultados, um deles que chamaremos de sucesso **(S)** e outro que chamaremos de fracasso **(F)**. À probabilidade de ocorrer sucesso em cada ensaio chamaremos de p; a probabilidade de fracasso chamaremos de q, de tal modo que q=1–p.


**Exemplo 1**: Qual é a probabilidade de obter face 5 em uma única tentativa do arremesso de um dado comum?

 **Resolução:**
Vamos afirmar que existe apenas 1 número 5 em um dado comum, para isso, temos que, x = 1.

A probabilidade de obtermos sucesso **(S)** é de 1/6, isso equivale a = 0.1666 ou 16.66%.

A probabilidade de obtermos fracasso **(F)** é de 5/6, isso equivale a = 0.8633 ou 86.34%.

 
#### Método de representar essa distribuição em forma de gráficos:

Usamos várias funções na biblioteca numpy para calcular matematicamente os valores de uma distribuição bernoulli. São criados histogramas sobre os quais plotamos a curva de distribuição de probabilidade

```python
dado_bernoulli = bernoulli.rvs(size=1000, p=0.6)
plt.title('Distribuição de Bernoulli')
ax = sns.distplot(dado_bernoulli,
                  kde=True,
                  color='crimson',
                  hist_kws={"linewidth": 25,'alpha':1})
ax.set(xlabel='Variáveis', ylabel='Frequência')
```

### 2. Distribuição Binomial

O modelo de distribuição binomial tem como objetivo encontrar a probabilidade de sucesso de um evento que possui apenas dois resultados possíveis em uma série de experimentos. Por exemplo, jogar uma moeda sempre dá uma cara ou coroa. A probabilidade de encontrar exatamente 3 caras jogando uma moeda repetidamente por 10 vezes é estimada durante a distribuição binomial.

**Exemplo 2:** Uma moeda não viciada é lançada várias vezes. Qual a probabilidade de que obtermos 5 caras antes de obtermos 3 coroas?

**Resolução:** Vamos considerar como sucesso a obtenção de cara em cada lançamento da moeda. Desta forma queremos obter 5 sucessos antes de obtermos 3 fracassos. Mas isto só é possível se jogarmos a moeda pelo menos 5 vezes e no máximo 7 vezes, pois em menos de 5 jogadas não é possível obtermos 5 caras e em 8 jogadas ou mais, já temos que ter obtido as 5 caras, pois caso contrário vamos ter obtido 3 coroas, ou mais o que não é o intuito.

Temos as variáveis aleatórias $ X_i\sim b(i,0,5) $, definidas como sendo o número de sucessos obtidos em $ i $ lançamentos da moeda com $ i = 5,6,7 $. Sendo assim precisamos calcular $ \mathbb{P}(X_i=5) $ para cada $ i $.

Portanto a probabilidade de obtermos 5 caras antes de 3 coroas é:

$ \mathbb{P}(X_5 =5)+\mathbb{P}(X_6=5)+\mathbb{P}(X_7=5)= \sum_{i=5}^7\left(\begin{array}{c} i\\ 5 \end{array} \right) (1/2)^5 (1-1/2)^{i-5}\thickapprox 0,23. $



#### Método de representar essa distribuição em forma de gráficos:

Usamos a biblioteca python transoceânica, que possui funções embutidas para criar esses gráficos de distribuição de probabilidade. Além disso, o pacote scipy ajuda a criar a distribuição binomial.<p>

```python
binom.rvs(size=10,n=20,p=0.8)

data_binom = binom.rvs(n=20,p=0.8,loc=0,size=1000)
ax = sns.distplot(data_binom,
                  kde=True,
                  color='blue',
                  hist_kws={"linewidth": 25,'alpha':1})
ax.set(xlabel='Valores', ylabel='Frequência')
plt.title('Distribuição Binomial')
plt.show()
```

### 3. Distribuição Poisson

Pode ser definida como uma distribuição de probabilidade de variável aleatória discreta que expressa a probabilidade de uma série de eventos ocorrer num certo período de tempo se estes eventos ocorrem independentemente de quando ocorreu o último evento.

**Exemplo 3:** Considere um processo que têm uma taxa de $ 0.2 $ defeitos por unidade. Qual a probabilidade de uma unidade qualquer apresentar: **dois defeitos?**

**Resolução:** Neste caso, temos que $ X \sim \ \text{Poisson}(\lambda) $ com $ \lambda=0.2 $. Então

$ \displaystyle\mathbb{P}(X=2)=\frac{e^{-0,2}(0,2)^2}{2!}=0.0164 $;

#### Método de representar essa distribuição em forma de gráficos:


```python
data_binom = poisson.rvs(mu=4, size=10000)
ax = sns.distplot(data_binom,
                  kde=True,
                  color='green',
                  hist_kws={"linewidth": 25,'alpha':1})
ax.set(xlabel='Valores', ylabel='Frequência')
plt.title('Distribuição Poisson')
```

### 4. Distribuição Geométrica
Essa distribuição é constituída por duas funções de probabilidade discretas:

 - a distribuição de probabilidade do número X de tentativas de Bernoulli necessárias para alcançar um sucesso, suportadas pelo conjunto { 1, 2, 3, ... }
 - a distribuição de probabilidade do número Y = X − 1 de insucessos antes do primeiro sucesso, suportadas pelo conjunto { 0, 1, 2, 3, ... }.

Para efetuar o cálculo dessa distribuição, temos a seguinte fórmula:
$ P(V=X) = S^X . F $ **(onde S é o sucesso de ocorrer a condição e F o fracasso).**

**Exemplo 4:** Considere o experimento em que uma moeda viciada é lançada sucessivas vezes, até que ocorra a primeira cara. Seja $ X $ a variável aleatória que conta o número de coroas obtidos no experimento (ou seja, a quantidade de lançamentos anteriores a obtenção da primeira cara). Sabendo que a probabilidade de cara é de $ 0.4 $, qual é a probabilidade de $ \mathbb{P}(2\leq X<4) $.

**Resolução:** 

$\mathbb{P}(2\leq X< 4)=\mathbb{P}(X=2)+\mathbb{P}(X=3)=0.6^2\cdot 0.4+0.6^3\cdot 0.4=0.2304.$

#### Método de representar essa distribuição em forma de gráficos:

Abaixo resultado gerado para contrastar em utilização com a fórmula em phyton scipy.stas.geom.pmf, que deve ser a mesma

```python
prob = 0.2
N = 12
p = np.zeros(N)
d = np.zeros(N)
for k in range(1,N+1):
    p[k-1] = (1-prob)**(k-1)*prob
    d[k-1] = geom.pmf(k,prob)
plt.bar(range(1,13),p, color = 'gray')
plt.xlim(1,13)
plt.xlabel('n')
plt.ylabel('P (n)')
plt.title('Distribuição Geométrica')
plt.show()

print('Probabilidade de sucesso depois de', N ,'tentativas é:',
      (1-sum(p)),'\nDe acordo com a fórmula.\n')
print('Probabilidade de sucesso depois de', N ,'tentativas é:',
      (1-sum(d)),'\nDe acordo com a funcao geom.pmf')
```

### 5. Distribuição Uniforme
A distribuição uniforme é a mais simples distribuição contínua, entretanto uma das mais importantes e utilizadas dentro da teoria de probabilidade. A distribuição uniforme tem uma importante característica a qual a probabilidade de acontecer um fenômeno de mesmo comprimento é a mesma.

**Exemplo 5:** A ocorrência de panes em qualquer ponto de uma rede telefônica de $ 7 $ km foi modelada por uma distribuição Uniforme no intervalo $ [0, 7] $. Qual é a probabilidade de que uma pane venha a ocorrer nos primeiros $ 800 $ metros?

**Resolução:** 

A função densidade da distribuição Uniforme é dada por $ f(x)=\frac{1}{7} $ se  $ 0\leq x\leq 7 $ e zero, caso contrário. 

Assim, a probabilidade de ocorrer pane nos primeiros 800 metros é:

$\mathbb{P}\left(X\leq 0,8\right)=\int_0^{0,8} f(x)dx=\frac{0,8-0}{7}=0,1142.$
 

#### Método de representar essa distribuição em forma de gráficos:

Devemos preencher com os parâmetros de distribuição desejados Depois de definir os parâmetros de distribuição dessa maneira, esses objetos de distribuição têm muitos métodos úteis, por exemplo:

dist.pmf(x) calcula a função de massa de probabilidade em valores x no caso de distribuições discretas

dist.pdf(x) calcula a função de densidade de probabilidade em valores x no caso de distribuições contínuas

dist.rvs(N) calcula Nvariáveis ​​aleatórias distribuídas de acordo com a distribuição fornecida

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.uniform(0,1,1000) # Gerando pontos 1000 pontos entre 0 e 1.
count, bins, ignored = plt.hist(data, 20, facecolor='green') 

plt.xlabel('X~U [0,1]')
plt.ylabel('Contagem')
plt.title("Distribuicao Uniforme")
plt.axis([0, 1, 0, 100]) # x_start, x_end, y_start, y_end
plt.grid(True)

plt.show(block = False)
```

### 6. Distribuição Exponencial

Podemos dizer que a distribuição Exponencial tem relação direta com a distribuição de Poisson. Esse modelo de distribuição de probabilidade descreve o tempo ou espaço entre dois sucessos consecutivos de uma variável de Poisson. Ou seja, exemplos do tipo (o tempo entre falhas de equipamentos ou o tempo entre chegadas de clientes a um shopping) são demonstrados pela **Distribuição Exponencial**.

#### Cálculo de Probabilidades na Distribuição Exponencial
É possível calcular as probabilidades da distribuição Exponencial utilizando a função densidade juntamente com uma integral definida. A função utilizada nesse cálculo é:
$ P(X>t) = e^λ.t $.

**Exemplo 6:** Suponha que, em determinado período do dia, o tempo médio de atendimento em um caixa de banco seja de 5 minutos. Admitindo que o tempo para atendimento tenha distribuição exponencial, determinar a probabilidade de um cliente: **esperar mais do que 5 minutos:**

**Resolução:**

Seja a variável aleatória X: tempo de atendimento.  Foi dado que o tempo médio de atendimento é de 5 minutos. Vimos que a média, ou esperança, de uma variável com distribuição exponencial é:
 
E(X) = 1/λ, com isso, temos: $1\frac{λ} = 5 => λ = 1/5 => **λ = 0,2**$ (Parâmetro da distribuição)

De acordo com a fórmula apresentada acima, temos:

**$ P(X>5) $** = e-0,2.5 = 0,3679 ou 36,79%, ou seja, a probabilidade do atendimento demorar 5 minutos é **36,79%**.

#### Método de representar essa distribuição em forma de gráficos:


```python
fig, ax = plt.subplots(1, 1)
mean, var, skew, kurt = expon.stats(moments='mvsk')

x = np.linspace(expon.ppf(0.01),
                expon.ppf(0.99), 100)

rv = expon()
ax.plot(x, rv.pdf(x), 'k-', lw=2, label='Distribuição Exponencial')

a=mean_squared_error(x, expon.cdf(x))
accuracy = 100-a

r = expon.rvs(size=1000)

ax.hist(r, density=True, histtype='stepfilled', alpha=0.2)
ax.legend(loc='best', frameon=False)

plt.xlabel('x')
plt.ylabel('Densidade (x)')

plt.show()
```

### 7. Distribuição Gaussiana

A distribuição normal ou Gaussiana, apresenta dados organizando a distribuição de probabilidade de cada valor nos dados. A maioria dos valores permanece em torno do valor médio, tornando o arranjo simétrico. 

Uma variável aleatória contínua $ X $ tem distribuição Normal se sua função densidade de probabilidade for dada por:

$ f(x)=\frac{1}{\sqrt{2\pi\sigma^2}}\exp\left[-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2\right], \quad x\in(-\infty,\infty). $


**Exemplo 7:** Calcular a área sob a curva para $ Z $ maior que $ 2,75 $.

**Resolução:** A área sob a curva normal para $ Z $ maior do que $ 2,75 $ é dada por


$\mathbb{P}(Z\geq 2,75)= \int_{2,75}^{\infty}\frac{1}{\sqrt{2\pi}}\exp\left[-\frac{x^2}{2}\right]=1-0,9970=0,003.$


Para achar a área sob a curva normal devemos conhecer dois valores numéricos, a média $ \mu $ e o desvio padrão $ \sigma $.

#### Método de representar essa distribuição em forma de gráficos:

Abaixo exemplo da função Gaussiana utilizando funções na biblioteca numpy para calcular matematicamente os valores de uma distribuição normal

```python
mu, sigma = 0.5, 0.1
s = np.random.normal(mu, sigma, 1000)

count, bins, ignored = plt.hist(s, 20, density=True)
plt.title('Distribuição Gaussiana')
plt.xlabel('x')
plt.ylabel('Densidade (x)')
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
    np.exp( - (bins - mu)**2 / (2 * sigma**2) ),linewidth=3, color='y')
plt.show()
```
